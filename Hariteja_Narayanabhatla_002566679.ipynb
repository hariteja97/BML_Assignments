```python
## heatmap for the columns and their dependencies.
Y=dataset['Spam']
x=dataset.drop(columns='Spam')
import seaborn as sns
coorelation_mat= x.corr().abs()
sns.heatmap(coorelation_mat, linewidths=0.6, annot=True, annot_kws={"fontsize":20})
plt.figure(figsize = (100,100))
sns.set(rc = {'figure.figsize':(60,60)})
plt.show()
## from the matrix we see that we have some columns which are highly correlated. Hence we need to remove those columns for machine to give
## good predicted values
```

```python
## we see that we have some columns which are correlated to other columns.. These type of columns reduce the model in giving
## good predicted value. we need to remove those columns so that we can have only a single column.
upper_tri = coorelation_mat.where(np.triu(np.ones(coorelation_mat.shape),k=1).astype(np.bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.50)] ## 0.5 is the threshold value for correlation
print(to_drop)

df1 = dataset.drop(columns=list(to_drop))

df1.head()

```

```python
##Now apply PCA on the data.
from sklearn.decomposition import PCA
from sklearn import preprocessing
pcs = PCA()
pcs.fit(x)
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                           'Proportion of variance': pcs.explained_variance_ratio_,
                           'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)})
pcsSummary_df.head(5)
```

```python
np.cumsum(pcs.explained_variance_ratio_)
```

```python
pcsSummary_df['Cumulative proportion']
len([i for i in list(pcsSummary_df['Cumulative proportion']) if i >0.99 ])
```

```python
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.head(5)
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), columns=pcsSummary_df.columns,index=x.columns)
pcsComponents_df.iloc[:,:5]
```
